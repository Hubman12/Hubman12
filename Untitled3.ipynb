```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.naive_bayes import GaussianNB
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('/Logistic_Iris.csv')
dataset
```

```python
x = dataset.iloc[:,[0,1,2,3]].values
x
```

```python
y = dataset.iloc[:,4].values
y
```

```python
#splitting the dataset into the training and testing set
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
Single = GaussianNB()
Single.fit(x_train,y_train)
```

```python
y_pred = Single.predict(x_test)
print("Prdicted values for single Naive bayes classifier:")
y_pred
```

```python
Acc_Single = accuracy_score(y_test,y_pred)*100
print("Accuracy of single Naive bayes classifier:",Acc_Single)
```

```python
cm = confusion_matrix(y_test,y_pred)
print("Confusion Matrix using single naive bayes classifier:",cm)
```

```python
num_class = 100
```

```python
bag = BaggingClassifier(estimator=Single,n_estimators=num_class,random_state=0)
bag.fit(x_train,y_train)
```

```python
results = model_selection.cross_val_score(bag,x_test,y_test,cv=10)
print("\n\nAccuracy using bagged set of naive bayes classifier:",results.mean()*100)
```

```python
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
x = dataset.iloc[:,[0,1,2,3]].values
x
```

```python
y = dataset.iloc[:,4].values
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
adaboost = AdaBoostClassifier(n_estimators=50,learning_rate=0.2)
adaboost.fit(x_train,y_train)
```

```python
adaboost.score(x_test,y_test)
```

```python
y_pred = adaboost.predict(x_test)
print("Predicted values for adaboost classifier:")
y_pred
```
